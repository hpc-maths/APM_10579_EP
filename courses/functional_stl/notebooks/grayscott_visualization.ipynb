# Grayscott visualization

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import matplotlib.cm as cm
```

```python
a = np.array(())
resized = False
with open('grayscott.txt') as f:
    nb_times = int(f.readline().strip('\n'))
    time_iter = 0
    matrix_size = 0
    while time_iter < nb_times:
        line = f.readline()
        line = list(map(int, f.readline().strip('\n').split(',')))
        matrix_size = line[0]
        if not resized:
            a.resize([nb_times] + line)
            resized = True
        for i in range(matrix_size):
            a[time_iter, i] = list(map(float, f.readline().strip(',\n').split(',')))
        time_iter = time_iter + 1
```

```python
def animate(i):
    image.set_array(a[i])
    return image,
        
```

```python
%matplotlib widget

fig = plt.figure()

image = plt.imshow(a[0], cmap = cm.Greys)

anim = animation.FuncAnimation(fig,animate, frames=500, interval=10)
#anim.save('grayscott.mp4', fps=20, dpi=300)

plt.show()
```
